#Operators in python

1. Arithmatic Operator

2. Relational Operator

3. Logical Operator

4. Bitwise Operator

5. Assignment Operator

6. Membership Operator




```python
#Arithmatic Operator
print(20+5)
print(20-5)
print(20*5)
print(20/5)
print(20%5)



#Relational Operator

print(20>5)
print(20<5)
print(20>=5)
print(20<=5)
print(20==5)
print(20!=5)

#Logical Operator
print("from this Logical Operator are printing")
print(True and True)
print(True and False)
print(False and True)
print(False and False)

print(True or True)
print(True or False)
print(False or True)
print(False or False)

print(not True)
print(not False)

print(1 and 0)
print(1 or 0)
print(not 1)
print(not 0)

```

```python
#Bitwise Operator

print(2&3)  #AND
print(2|3)  #OR
print(2^3)  #Ex OR
print(~2)   #not
print(2>>3)
print(2<<3)
```

```python
#Assignment Operator
a=2
print(a)
a+=2
print(a)
a-=2
print(a)
a*=2
print(a)
a/=2
print(a)
a%=2
print(a)
```

```python
#Membership Operator
print(2 in [1,2,3,4,5])
print(2 not in [1,2,3,4,5])
print("a" in "Suraj")
print("a" not in "Suraj")
print('A' in 'Suraj saw')
print('A' not in 'Suraj saw')  #case sensitivity
```

```python
#gessing game

#generate a random no. from 1 to 100

#we use random library

import random

jackpot= random.randint(1 , 100)
guess=int(input("guess karo : "))
counter=1
while guess!=jackpot:
  if guess<jackpot:
    print("Worng hai ! guess higher no. ")
  else :
    print("Worng hai ! guess lower no. ")
  guess=int(input("guess karo : "))
  counter+=1
else:
  print("Correct guess : ",jackpot)
  print("Attempts",counter)
```

Program - The current population of a town is 10000. The population of the town is
increasing at the rate of 10% per year. You have to write a program to find out the
population at the end of each of the last 10 years.

```python
current_population=10000
print("Current population : ",current_population)
for i in range(10):
  current_population =  current_population + current_population*0.1
  print("population of ",i +2025, "year :",current_population)


```
