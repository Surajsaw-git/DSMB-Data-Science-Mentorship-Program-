#LOOP Control Statement
* Break
* Continue
* Pass


```python
for i in range(1,10):
  if i==5:
    break
  print(i)
```

USE OF BREAK
* Searching in Database
* Break: This statement terminates the loop entirely and resumes execution in the statement immediately following the loop.
* Break: You might use break to exit a search loop once the desired item is found.

```python
lower=int(input("Enter the value of lower range :"))
upper=int(input("Enter the value of upper range :"))

for i in range(lower,upper+1):
  for j in range(2,i):
    if i%j==0:
      break
  else:
    print(i)
```

#Continue
* Continue: This statement skips the rest of the current iteration of the loop and moves to the next iteration.
* Continue: You might use continue to skip processing invalid data in a data processing loop.
#USE
* E commerce site : if the product is out of stock then , we have to skip the product.


```python
for i in range(1,10):
  if i==5:
    continue
  print(i)
```

# Pass
* Pass: This statement does nothing and is used as a placeholder where a statement is syntactically required but you don't want to execute any code.
* Pass: You might use pass when defining a loop structure for future implementation or to avoid syntax errors in incomplete code.


```python
for i in range(1,10):
```

```python
for i in range(1,10):
  pass
```

Write a program that can check whether a given string is palindrome or not.

```python
#palindrom
#abba,malayalam etc
s=input("Enter Any string : ")
flag=True
for i in range(0,len(s)//2):
  if s[i] != s[len(s)-i-1]:
    flag = False
    break
if flag==True:
  print("The given string is palindrome")
else:
  print("The given string is not palindrome")

```

write a program to count the number of words in a string without split().

```python

s=input("Enter Any string : ")
temp = ''
L=[]
for i in s:
  if i != ' ':
    temp = temp + i
  else:
    L.append(temp)    #append used for add value in string
    temp=''
L.append(temp)
print(L)
print(len(L))
```

Write a python program to convert a string to title case without use the title().

```python
s=input("Enter Any string : ")
L=[]
for i in s.split():

  L.append(i[0].upper() + i[1:].lower())
print(L)
' '.join(L)
```

Write a python program that can convert integer to string.

```python
from string import digits
number=int(input("Enter a number : "))
digits = '0123456789'
result = ''
while number !=0:
  result = digits[number % 10] + result
  number = number//10
print(result)
print(type(result))
```
