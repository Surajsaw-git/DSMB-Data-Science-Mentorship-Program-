##List Comprehension
List comprehension provides a concise way of creating  lists.

newlist=[expresion for item in iterable if condition == true]

Advantages of List Comprehension

* More time-efficient and space-efficient than loops.
* Require fewer lines of code.
* Transforms iterative statement into a formula.

```python
#Add 1 to 10 numbers to a list
L=[]
for i in range(1,11):
  L.append(i)
print(L)
```

```python
L = [i for i in range(1,11)]
print(L)
```

```python
# Scalar multiplication on a vector
v=[2,3,4]
s = -3
#[-6,-9,-12]
print(v*s)

x=[]
for i in v:
  x.append(i*s)
print(x)
```

```python
y=[s*i for i in v]
print(y)
```

```python
# Add square

L=[1,2,3,4,5]
x=[i**2 for i in L]
print(x)
```

```python
#print all the number from 1 to 100 which is divisible by 5
x=[i for i in range(1,101) if i%5==0]
print(x)
```

```python
#find the languages which start with letter p

languages=['python','java','css','php','r','c','c++','react']
x=[language for language in languages if language.startswith('p')]
print(x)
```

```python
#find the languages which start with letter p

languages=['python','java','css','php','r','c','c++','react']
[language for language in languages if language.startswith('p')]
```

```python
# Nested if with List Comprehension
basket = ['apple' , 'guava' , 'cherry' , 'banana' ]
my_fruits = ['apple' , 'kiwi' , 'grapes' , 'banana' ]
```

```python
# add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'
[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
```

```python
# Print a (3,3) matrix using list comprehension -> Nested List comprehension
[[i*j for i in range(1,4)] for j in range(1,4)]
```

```python
#cartesian prooduct

L1=[1,2,3,4,5]
L2=[6,7,8,9]

[i*j for i in L1 for j in L2]
```

##Zip

The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together,
and then the second item in each passed iterator are paired together.

If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator.

```python
# Write a program to add items of 2 lists indexwise

L1 = [1,2,3,4]
L2 = [-1,-2,-3,-4]

list(zip(L1,L2))

[i+j for i,j in zip(L1,L2)]
```

```python
# Write a program to add items of 2 lists indexwise

L1 = [1,9,13,84,8]
L2 = [-8,-2,-3,-4,2]

list(zip(L1,L2))

[i+j for i,j in zip(L1,L2)]
```

```python
L1=[2,4,5]
L2=[5,4,6]
L1.reverse()
L2.reverse()
list(zip(L1,L2))
[i+j for i,j in zip(L1,L2)]
```
