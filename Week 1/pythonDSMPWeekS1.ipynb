#compiler vs interpreter
Compilers and interpreters are both tools that translate human-readable programming code into machine-executable code, but they differ in how they handle the translation process. Compilers translate the entire program at once into machine code, creating an executable file, while interpreters translate and execute code line by line during runtime.

#compiler

###Translation Process:
Translates the entire source code into machine code before execution, generating an executable file.
###Error Detection:
Reports errors after the entire program has been scanned.
###Execution Speed:
Generally faster execution because the translation is done upfront.
###Memory Usage:
May require more memory as it generates and stores the executable file.
###Example Languages:
C, C++, Java.

##Interpreters

###Translation Process:
Translates and executes code line by line during runtime, without generating an executable file.
###Error Detection:
Reports errors as they are encountered during execution.
###Execution Speed:
Generally slower execution compared to compiled code because translation happens during runtime.
###Memory Usage:
Requires less memory as it doesn't create a separate executable file.
###Example Languages:
Python, Ruby, JavaScript.
