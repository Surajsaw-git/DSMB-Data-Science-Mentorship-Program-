# Tuples

A tuple in Python is similar to a list. The difference between the two is that we cannot change the elements of a tuple once it is assigned whereas we can change the elements of a list.

In short, a tuple is an immutable list. A tuple can not be changed in any way once it is created.

Characterstics

- Ordered
- Unchangeble
- Allows duplicate

### Plan of attack

- Creating a Tuple
- Accessing items

- Adding items

- Operations on Tuples
- Tuple Functions

```python
t1=()
print(t1)

# create a tuple with single item
t2=('suraj')
print(type(t2))

#this is a single item tuple
t3=('suraj',)
print(t3)
print(type(t3))
```

```python
T=(3,2,1,5,4,5)
print(T)
print(type(T))

T=(3,(8,9,7,4,5),1,5,4,5)
print(T)
print(type(T))

T=(((4,5,9,6,3),(8,9,7,4,5)),((4,5,6,7),(1,6,4,5,2)))
print(T)
print(type(T))
```

### Difference between Lists and Tuples

- Syntax
- Mutability
- Speed
- Memory
- Built in functionality
- Error prone
- Usability

```python
import time

L = list(range(100000000))
T = tuple(range(100000000))

start = time.time()
for i in L:
  i*5
print('List time',time.time()-start)

start = time.time()
for i in T:
  i*5
print('Tuple time',time.time()-start)
```

```python
import sys

L = list(range(1000))
T = tuple(range(1000))

print('List size',sys.getsizeof(L))
print('Tuple size',sys.getsizeof(T))
```

#Sets

A set is an unordered collection of items. Every set element is unique (no duplicates) and must be immutable (cannot be changed).

However, a set itself is mutable. We can add or remove items from it.

Sets can also be used to perform mathematical set operations like union, intersection, symmetric difference, etc.

Characterstics:

* Unordered

* Mutable

* No Duplicates
* Can't contain mutable data types



###Creating Sets

```python
#empty
s={}               #this is not an empty set, it is create empty dict
print(s)
print(type(s))


#how to create an empty set
s=set()               #this is an empty set,
print(s)
print(type(s))
```

```python
#1D and 2D
s1={3,4,5,1,9}     #this is also a homogenous set (same data type int)
print(s1)
print(type(s1))
```

```python
#2D set
s1={3,4,5,1,{9}}
print(s1)
print(type(s1))

#2D sets are not possible
```

```python
# hetro
s1={"suraj",3,9.3}
print(s1)

# hetro
s1={"suraj",3,9.3,(1,2,3)}
print(s1)
```

```python
#we can use type consersion
s2=set([1,2,3,4,5])
print(s2)
print(type(s2))
```

```python
# Duplicates are not allow in ssets
s1={1,2,3,5,8,9,8,9,1}
print(s1)
```

```python
# sets are Unordered
s1={1,2,3}
s2={3,2,1}
print(s1==s2)
```

## Sets Operations

```python
# Union (|)

s1={1,2,3,4,5,6}
s2={4,5,6,7,8,9,10}

print(s1|s2)
```

```python
# Insection (&)
s1={1,2,3,4,5,6}
s2={4,5,6,7,8,9,10}
print(s1&s2)
```

```python
# Difference (-)
s1={1,2,3,4,5,6}
s2={4,5,6,7,8,9,10}
print(s1-s2)

s1={1,2,3,4,5,6}
s2={4,5,6,7,8,9,10}
print(s2-s1)
```

```python
# Symmetric Difference (^)
s1={1,2,3,4,5,6}
s2={4,5,6,7,8,9,10}
print(s1^s2)
```

##Set Functions
* len
* sum
* max
* min
* sorted


```python

s2={4,5,6,7,8,9,10}
print(len(s2))

print(sum(s2))

print(max(s2))

print(min(s2))

print(sorted(s2))   # when we sort any data sets then it will store in list
```

```python
# Union / union update

s1={1,2,3,4,5,6}
s2={4,5,6,7,8,9,10}

print(s1.union(s2))
s1.update(s2)
print(s1)
print(s2)
```

```python
# intersection / intersection_update
s1={1,2,3,4,5,6}
s2={4,5,6,7,8,9,10}

s1.intersection(s2)
```

```python
# intersection / intersection_update
s1={1,2,3,4,5,6}
s2={4,5,6,7,8,9,10}

s1.intersection_update(s2)
print(s1)
print(s2)
```

```python
# Difference / Difference update

s1={1,2,3,4,5,6}
s2={4,5,6,7,8,9,10}

print(s1.difference(s2))
s1.difference_update(s2)
print(s1)
print(s2)
```

```python
# Symmetric / symmetric_difference_update

s1={1,2,3,4,5,6}
s2={4,5,6,7,8,9,10}

print(s1.symmetric_difference(s2))
s1.symmetric_difference_update(s2)
print(s1)
print(s2)
```

```python
# isdisjoint/issubset/issuperset

s1={1,2,3,4,5,6}
s2={4,5,6,7,8,9,10}

s1.isdisjoint(s2)
```
