# Tuples

A tuple in Python is similar to a list. The difference between the two is that we cannot change the elements of a tuple once it is assigned whereas we can change the elements of a list.

In short, a tuple is an immutable list. A tuple can not be changed in any way once it is created.

Characterstics

- Ordered
- Unchangeble
- Allows duplicate

### Difference between Lists and Tuples

- Syntax

```python
import time

L = list(range(100000000))
T = tuple(range(100000000))

start = time.time()
for i in L:
  i*5
print('List time',time.time()-start)

start = time.time()
for i in T:
  i*5
print('Tuple time',time.time()-start)
```

```python
import sys

L = list(range(1000))
T = tuple(range(1000))

print('List size',sys.getsizeof(L))
print('Tuple size',sys.getsizeof(T))
```

#Sets

A set is an unordered collection of items. Every set element is unique (no duplicates) and must be immutable (cannot be changed).

However, a set itself is mutable. We can add or remove items from it.

Sets can also be used to perform mathematical set operations like union, intersection, symmetric difference, etc.

Characterstics:

* Unordered

* Mutable

* No Duplicates
* Can't contain mutable data types



###Creating Sets

```python
#empty
s={}               #this is not an empty set, it is create empty dict
print(s)
print(type(s))


#how to create an empty set
s=set()               #this is an empty set,
print(s)
print(type(s))
```

```python
#1D and 2D
s1={3,4,5,1,9}     #this is also a homogenous set (same data type int)
print(s1)
print(type(s1))
```

```python
#2D set
s1={3,4,5,1,{9}}
print(s1)
print(type(s1))

#2D sets are not possible
```

```python
# Union / union update

s1={1,2,3,4,5,6}
s2={4,5,6,7,8,9,10}

print(s1.union(s2))
s1.update(s2)
print(s1)
print(s2)
```

```python
# Symmetric / symmetric_difference_update

s1={1,2,3,4,5,6}
s2={4,5,6,7,8,9,10}

print(s1.symmetric_difference(s2))
s1.symmetric_difference_update(s2)
print(s1)
print(s2)
```

It looks like youâ€™ve run into a common Python "gotcha."
The error TypeError: 'dict' object is not callable usually happens because the name dict has been accidentally overwritten earlier in your code.
##What happened?
At some point in your current session, you likely assigned a variable named dict, like this:
* dict = { ... }
When you do that, Python forgets that dict() is a built-in function and thinks it is now just a specific dictionary. So, when you try to call dict([...]), Python complains because you can't "call" a dictionary like a function.

##How to fix it
You need to delete your custom variable to let Python "see" the built-in function again. You can do this by running:

python
* del dict

```python
#using sequence and dict
del dict
d2 = dict([(1,1),(2,2),(3,3)])
d2
```

##Dictionary Comprehension

{key: value for vars in iterable}

```python
# Print 1st 10 numbers and their squares

{i:i*i for i in range(1,11)}
```

```python
# using exiting dict

distance={'delhi':1000,'pune':2000,'mumbai':3000}
{key:value*0.62 for (key,value) in distance.items()}
```

```python
distance={'delhi':1000,'pune':2000,'mumbai':3000}
distance.items()
```

```python
#using Zip

days=["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
tem_c=[32.5,29.5,25.41,15.15,42.23,12.52,32.21]

{i:j for (i,j) in zip(days,tem_c)}
```

```python
# using if condition

product = {'phone':10,'laptop':0,'charger':32,'tablet':0}
{
    key:value for (key,value) in product.items() if value>0
}
```

```python
#Nested Comprehension

#print tables of number from 2 to 4

{i:{j:i*j for j in range(1,11)} for i in range(2,5)}
```
